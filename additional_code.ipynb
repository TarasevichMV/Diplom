```python
import pandas as pd
from catboost import Pool, CatBoostClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_excel("Анализ стратегий конкурентов.xlsx")
```

```python
data = data.drop(columns = ["Unnamed: 0"])
data = data.drop(0)
data = data.drop(536)
```

```python
data
```

```python
df = data.drop(
    columns = ["Новости",
               "В презентацию",
               "Компания",
               "Месяц",
               "Год",
               "Дата",
               "Нефтегаз",
               "Энергетика",
               "Строительство",
               "Машиностроение",
               "Прочие",
               "Ссылка"]
    )
```

```python
df = df.fillna(0)
df = df.replace(["+"], ["1"])
```

```python
df
```

```python
texts = df["Текст новости"]
categories = list(df.columns)
categories.pop(-1)
```

```python
categories
```

```python
dfs = [pd.DataFrame({"text": texts, category: df[category]}) for category in categories]
```

```python
dfs_split = [train_test_split(df.text, df[df.columns[1]], test_size=0.2, random_state=42) for df in dfs]
```

```python
def fit_catboost(X_train, X_test, y_train, y_test, catboost_params={}, verbose=100):
    learn_pool = Pool(
        X_train,
        y_train,
        text_features=["text"],
        feature_names=list(X_train)
    )
    test_pool = Pool(
        X_test,
        y_test,
        text_features=["text"],
        feature_names=list(X_train)
    )

    catboost_default_params = {
        'iterations': 50,
        'learning_rate': 0.03,
        'eval_metric': 'Accuracy',
    }

    catboost_default_params.update(catboost_params)

    model = CatBoostClassifier(**catboost_default_params)
    model.fit(learn_pool, eval_set=test_pool, verbose=verbose)

    return model
```

```python
models = {}
for data in dfs_split:
  model = fit_catboost(pd.DataFrame(data[0]), pd.DataFrame(data[1]), pd.DataFrame(data[2]), pd.DataFrame(data[3]))
  models[data[2].name] = model
```
